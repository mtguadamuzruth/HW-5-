```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_studyData = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
combined_studyData.head()
```

```python
# Checking the number of mice.
len(combined_studyData["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
combined_studyData.duplicated(subset=['Mouse ID', 'Timepoint'])
duplicate_mice = combined_studyData.loc[combined_studyData.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
Duplicate_mice = combined_studyData[combined_studyData.duplicated(subset=['Mouse ID', 'Timepoint'])]
Duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_combined_studyData= combined_studyData[combined_studyData['Mouse ID'].isin(duplicate_mice)==False]
clean_combined_studyData
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_combined_studyData["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean_regimen = clean_combined_studyData.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
mean_regimen
median_regimen = clean_combined_studyData.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
median_regimen
var_regimen = clean_combined_studyData.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
var_regimen
std_regimen = clean_combined_studyData.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
std_regimen
SEM_regimen = clean_combined_studyData.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
SEM_regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
single_summary = pd.DataFrame({"mean": mean_regimen,"median": median_regimen,"variance": var_regimen,"standard deviation": std_regimen, "SEM": SEM_regimen})
single_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
agg_function = {
    'Tumor Volume (mm3)': ["mean","median","var", "std", "sem"],
}

clean_combined_studyData.groupby('Drug Regimen').agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
```
